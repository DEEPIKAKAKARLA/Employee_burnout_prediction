```python
import numpy as np
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import os

def generate_art(size=(512, 512), num_shapes=100, save_path='generative_art'):
    os.makedirs(save_path, exist_ok=True)

    # Create a new image with white background
    image = Image.new('RGB', size, 'white')
    draw = ImageDraw.Draw(image)

    for _ in range(num_shapes):
        shape_type = np.random.choice(['circle', 'rectangle', 'line'])
        color = tuple(np.random.randint(0, 255, size=3))  # Random color

        x1, y1 = np.random.randint(0, size[0]), np.random.randint(0, size[1])
        x2, y2 = np.random.randint(0, size[0]), np.random.randint(0, size[1])

        if shape_type == 'circle':
            radius = np.random.randint(5, 50)
            draw.ellipse([(x1-radius, y1-radius), (x1+radius, y1+radius)], outline=color, width=3)

        elif shape_type == 'rectangle':
            draw.rectangle([(x1, y1), (x2, y2)], outline=color, width=3)

        elif shape_type == 'line':
            draw.line([(x1, y1), (x2, y2)], fill=color, width=3)

    # Save image
    image.save(os.path.join(save_path, 'generative_art.png'))

    # Optional: Display image
    plt.imshow(image)
    plt.axis('off')
    plt.show()

if __name__ == "__main__":
    generate_art()
```
